```python
BASE = "https://api.openalex.org/authors?"
ENDPOINT = "authors?"
API = "api_key=CIcafSHBHxopqOewXo9hYK"

url = BASE + API

names_series = Navne.iloc[:, 0].astype(str)

rows = []

no_results = 0
errors = 0

for name in names_series:
    params = {
        "search": name,
        "per-page": 200,
        "page": 1
    }

    data = requests.get(url, params=params).json()

    results = data.get("results", [])

    if not results:
        continue

    n = results[0] 
    
    h_index = (n.get("summary_stats") or {}).get("h_index") 
    
    country = (
        n["affiliations"][0]["institution"]["country_code"]
        if n["affiliations"] else None
    )

    rows.append({
        "id": n["id"],
        "display_name": n["display_name"],
        "works_count": n["works_count"],
        "work_api": n["works_api_url"],
        "country_code": country,
        "h_index": h_index
    })

df_d1 = pd.DataFrame(rows).drop_duplicates("id")
df_d1.to_csv("df_d1.csv", index=False)
df_d1
```

This ensures you can systematically retrieve the complete set of works beyond the initial 200.
```python
df_d1 = pd.read_csv("../Week 2/df_d1.csv")

df_d1 = df_d1[(df_d1["works_count"] >= 5) & (df_d1["works_count"] <= 5000)]
df_d1
```

```python
rows_d2_3 = []

api = "CIcafSHBHxopqOewXo9hYK"

for name in df_d1["work_api"]:
    
    page = 1
    
    
    while True:
    
        url = (
        f"{name},authors_count:<10,cited_by_count:>10,"
        "concepts.id:C144024400|C15744967|C162324750|C17744445,"
        "concepts.id:C33923547|C121332964|C41008148"
        f"&per-page=200&page={page}"
        "&select=id,publication_year,cited_by_count,title,abstract_inverted_index,authorships"
        f"&api_key={api}"        )

        data = requests.get(url).json()
        results = data.get("results", [])

        if not results:
            break

        for work in results:
            corresponding_ids = []
            for authorship in work["authorships"]:
                    corresponding_ids.append(authorship["author"]["id"])
            
            rows_d2_3.append({
                "work_id": work["id"],
                "title": work["title"],
                "publication_year": work["publication_year"],
                "cited_by_count": work["cited_by_count"],
                "author_ids": corresponding_ids,
                "abstract_inverted_index": work["abstract_inverted_index"]
            })
            
        if len(results) < 200:
            break
            
        page += 1  

df2_3 = pd.DataFrame(rows_d2_3).drop_duplicates("work_id")
df2_3
```

```python
df_2 = df2_3[[
    "work_id",
    "publication_year",
    "cited_by_count",
    "author_ids"
]].copy()

df_2.to_csv("D_2.csv", index=False)

df_3 = df2_3[[
    "work_id",
    "title",
    "author_ids"
]].copy()

df_3.to_csv("D_3.csv", index=False)
```
